```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
import numpy as np
import pandas as pd
```

```python
df = sns.load_dataset('titanic')
```

```python
df.head()
```

```python
# create a count plot
a = sns.countplot(data =df ,x = 'survived')
plt.title('Count of survival of people')
a.set_xticks(range(2))
a.set_xticklabels(['Died', 'Survived'])
a.set_ylabel("No. of people")
```

```python
# create a box plot
b = sns.boxplot(x = 'survived', y = 'age',hue = 'survived', data =df)
b.set_xticks(range(2))
b.set_xticklabels(['Died', 'Survived'])
b.set_ylabel("No. of people")
```

```python
# create a violin plot
c = sns.violinplot(x = 'pclass', y = "fare",hue = "pclass", data = df)
```

```python
# create a histogram
d = sns.histplot(data = df,kde = True, x = 'age')
```

```python
fig, ax = plt.subplots(2, 2, figsize=(20,20) )
a = sns.countplot(data =df ,x = 'survived',ax=ax[0,0])
a.set_title('Count of survival of people')
ax[0,0].set_xticks(range(2))
ax[0,0].set_xticklabels(['Died', 'Survived'])
ax[0,0].set_ylabel("No. of people")
b = sns.boxplot(x = 'survived', y = 'age',hue = 'survived', data =df,ax = ax[0,1])
b.set_title('box plot')
b.set_xticks(range(2))
b.set_xticklabels(['Died', 'Survived'])
b.set_ylabel("No. of people")
c = sns.violinplot(x = 'pclass', y = "fare", data = df, ax=ax[1,0])
c.set_title('fare w.r.t passengre class')
d = sns.histplot(data = df,kde = True, x = 'age',ax = ax[1,1])
d.set_title('AGE')
```
