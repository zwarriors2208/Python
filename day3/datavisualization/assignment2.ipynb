```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
months = ['January', 'February', 'March', 'April', 'May', 'June', 'July', 'August', 'September', 'October', 'November', 'December']
```

```python
df = pd.DataFrame(np.random.randint(50, 500,(12,5)))
df.index = ['January', 'February', 'March', 'April', 'May', 'June', 'July', 'August', 'September', 'October', 'November', 'December']
df.columns = ['product 1','product 2','product 3','product 4','product 5',]
```

```python
sns.lineplot(x = df.index, y = 'product 1', data = df, hue = df.index)
sns.lineplot(x = df.index, y = 'product 2', data = df)
sns.lineplot(x = df.index, y = 'product 3', data = df)
sns.lineplot(x = df.index, y = 'product 4', data = df)
sns.lineplot(x = df.index, y = 'product 5', data = df)
```

```python
df.plot(kind = 'bar', stacked = True)
```

```python
a1 = np.random.randint(50, 500,(12,))
a2 = np.random.randint(50, 500,(12,))
a3 = np.random.randint(50, 500,(12,))
a4 = np.random.randint(50, 500,(12,))
a5 = np.random.randint(50, 500,(12,))
```

```python
fig, ax = plt.subplots(figsize=(10, 5))
sns.barplot( x = months, y = a1, ax = ax, hue = a1)
sns.barplot( x = months, y = a2,  ax = ax, hue = a2)
sns.barplot( x = months, y = a3,  ax = ax, hue = a3)
sns.barplot( x = months, y = a4, ax = ax, hue = a4)
sns.barplot( x = months, y = a5, ax = ax, hue = a5)
plt.legend(df.columns)
```
