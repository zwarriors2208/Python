```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
day = np.arange(1,21)
```

```python
temp = np.random.uniform(15,30, size = 20)
```

```python
humid = np.random.randint(40,90, (20))
humid
```

```python
df = pd.DataFrame({'days': day, 'temp':temp, 'humidity':humid})
df.head()
```

```python
sns.lineplot(x = 'days', y = 'temp', data = df)
```

```python
sns.scatterplot(x = 'temp', y = 'humidity', data = df,hue = 'days', s = 200)
```

```python
fig, ax = plt.subplots(figsize = (8,6))
sns.boxplot(x = 'days', y = 'temp', data = df,hue = 'humidity', palette='Set3',ax = ax)
# sns.boxplot(data = df)
plt.show()
```

```python
sns.histplot(x = 'humidity',data=df)
plt.show()
```

```python
fig, ax = plt.subplots(figsize=(12, 6))
sns.histplot(df, x='humidity', hue='days', multiple='stack', palette='inferno', ax=ax)
```
