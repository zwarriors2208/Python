```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
x = np.array(['January', 'February', 'March', 'April', 'May', 'June', 'July', 'August', 'September', 'October', 'November', 'December'])
```

```python
df = pd.DataFrame(np.random.randint(50, 500,(12,5)))
df.index = ['January', 'February', 'March', 'April', 'May', 'June', 'July', 'August', 'September', 'October', 'November', 'December']
df.columns = ['product 1','product 2','product 3','product 4','product 5',]
```

```python
df
```

```python
a1 = np.random.randint(50, 500,(12,))
a2 = np.random.randint(50, 500,(12,))
a3 = np.random.randint(50, 500,(12,))
a4 = np.random.randint(50, 500,(12,))
a5 = np.random.randint(50, 500,(12,))
```

```python
plt.figure(figsize=(12,6))
plt.plot(x,a1,label='Product 1',color= 'blue', marker='o',linestyle= '-')
plt.plot(x,a2,label='Product 2',color= 'orange', marker='s',linestyle= '--')
plt.plot(x,a3,label='Product 3',color= 'red', marker='*',linestyle= ':')
plt.plot(x,a4,label='Product 4',color= 'purple', marker='d',linestyle= '-.')
plt.plot(x,a5,label='Product 5',color= 'green', marker='h',linestyle= '-')
plt.title('Sales Analysis')
plt.xlabel('X axis')
plt.ylabel('Y axis')
plt.legend()
plt.grid(True)
plt.show()
```
